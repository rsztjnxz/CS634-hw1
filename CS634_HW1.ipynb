```python
m=10000
N=[1,5,10,20]
import numpy as np
from matplotlib import pyplot as plt
results=np.zeros((len(N),m))
print(results)
for i in range(len(N)):
    for j in range(m):
      results[i,j]=np.mean(np.random.uniform(0,1,N[i]))
for i in range(len(N)):
    plt.hist(results[i],bins=60,color = 'green')
    plt.show()
```

# Question 2(2)

```python
mean=[0,2]
cov=[[0.3,-1],[-1,5]]
pts=np.random.multivariate_normal(mean,cov,size=10000)
pts.mean(axis=0)
np.cov(pts.T)
plt.plot(pts[:,0],pts[:,1],',',alpha=0.5)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

plt.style.use('seaborn-dark')
plt.rcParams['figure.figsize']=14,6

random_seed=1000

cov_cal=[0.3,-1,-1,5]

mean=np.array([0,2])

for idx, val in enumerate(cov_cal):
    plt.subplot(1,3,idx+1)

    cov=np.array([[1,val],[val,1]])

    distr=multivariate_normal(cov=cov,mean=mean,
                              seed=random_seed)
    
    data=distr.rvs(size=5000)

    plt.plot(data[:,0],data[:,1],'o',c='lime',
             markeredgewidth=0.5,
             markeredgecolor='black')
    plt.title('Covariance between x1 and x2 = {val}')
    plt.xlabel('x1')
    plt.ylable('x2')
    plt.axis('equal')

plt.show()

```

```python
from IPython.lib.security import random
n=[10,100,1000]
mu=0
sigma=0.3
def simulate_condition(n,mu,sigma):
  vec=[]
  for i in range(n):
    x = mu + sigma[i,i] * random()
    vec.append(x)
  return vec

def plot_condition(vec,mu,sigma):
  plt.plot(vec,mu,'b')
  plt.plot(vec,sigma[:,0],'r')
  plt.title('Condition Distribution')
  plt.xlabel('Sample')
  plt.ylabel('Value')   
```
